```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

```python
img = 'https://media.product.which.co.uk/prod/images/original/gm-9ffc530c-f7af-454f-9f9c-57a000c62d3a-traffic-jamlead.jpeg'
```

```python
results = model(img)
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
results.xyxy
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    if frame is not None:  # Check if frame is valid
        results = model(frame)
    
        cv2.imshow('YOLO', np.squeeze(results.render()))
    
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    else:  # Exit loop if frame is invalid
        break

cap.release()
if cv2.getTrackbarPos('Quit', 'YOLO') == 1:
    cv2.destroyAllWindows()

```
